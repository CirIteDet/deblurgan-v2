```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
cap = cv2.VideoCapture('sp-AnUgcfzk_1_slow_1.mp4')
cap_deblur = cv2.VideoCapture('sp-AnUgcfzk_1_slow_1_deblur.mp4')
```

```python
i = 1
while(i<10):
    i += 1
    ret, frame = cap.read()
    ret_deblur, frame_deblur = cap_deblur.read()
    if ret == True and ret_deblur == True:
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame_deblur = cv2.cvtColor(frame_deblur, cv2.COLOR_BGR2RGB)
        gap = frame - frame_deblur
        plt.figure()
        plt.imshow(frame)
        plt.show()
        print('blur')
        plt.figure()
        plt.imshow(frame_deblur)
        plt.show()
        print('deblur')
    else:
        break
        

```
